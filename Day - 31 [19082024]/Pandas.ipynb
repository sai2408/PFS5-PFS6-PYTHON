# loc and iloc

```python
data = pd.Series(['a','b','c','d'], index = [1,3,5,7])
print(data)
print(data.loc[1])
print(data.iloc[1])
```

```python
data = pd.Series(['a','b','c','d'], index = [1,3,5,7])
print(data)
print('**************************************')

print('loc values') # it will take manually generated index values
print(data.loc[1])
print(data.loc[3])
print(data.loc[5])
print(data.loc[7])
print('**************************************')

print('iloc values') # it will take original index values
print(data.iloc[0])
print(data.iloc[1])
print(data.iloc[2])
print(data.iloc[3])
print('**************************************')
print(data.loc[1:7])
print(data.iloc[0:3])
```
